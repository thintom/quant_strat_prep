```python
# --- Problem: Group Anagrams ---
# Logic: Use a Hash Map to group strings by their sorted 'signature'.
# Time Complexity: O(n * m log m) where n is number of strings, m is avg length.
# Space Complexity: O(n * m) to store the grouped strings.

def groupAnagrams(strs: list[str]) -> list[list[str]]:
    res = {}
    for elem in strs:
        a = ''.join(sorted(elem))
        if a in res:
            res[a].append(elem)
        else:
            res[a] = [elem]
    return list(res.values())

# Test Case
example = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(f"Result: {groupAnagrams(example)}")
```

```python
# Identifying the top k elements from a given list
# run time o(n + D logD + D*K)
# space o(n)

def topKFrequent(nums: list[int], k: int) -> list[int]:
    res = {}
    for elem in nums:
        if elem in res:
            res[elem] += 1
        else:
            res[elem] = 1

    res_sort = sorted(list(res.values()), reverse = True)

    res_final = []

    for elem in res:
        if res[elem] in res_sort[:k]:
            res_final.append(elem)
    
    return res_final


# Better solution ie faster
# run time o(n + D logD)

def topKFrequent(nums: list[int], k: int) -> list[int]:
    res = {}
    for elem in nums:
        if elem in res:
            res[elem] += 1
        else:
            res[elem] = 1

    res_sort = [item[0] for item in sorted(res.items(), key=lambda x: x[1], reverse=True)][0:k]
    
    return res_sort[0:k]

# Even faster solution
# run time o(n)
def topKFrequent(nums: list[int], k: int) -> list[int]:
    counts = {}

    for elem in nums:
        counts[elem] = counts.get(elem, 0) + 1

    n = len(nums)
    buckets = [[] for _ in range(n + 1)]

    for a,b in counts.items():
        buckets[b].append(a)
        
    result = []

    for i in range(n , 0, -1):
        result += buckets[i]
        if len(result) >= k:
            break
    
    return result[0:k]

# Test Case
example = [1,1,1,2,2,3]
k = 2
print(f"Result: {topKFrequent(example, 2)}")
```

```python
# sorting algorithm
# O(nlogn)
def merge_list(lst1 : list[int], lst2 : list[int]) -> list:
    res = []
    if lst1[0] >= lst2[0]:
        if len(lst1) > 1:
            return lst1[0:1] + merge_list(lst1[1:], lst2)
        else:
            return lst1[0:1] + lst2
    else:
        if len(lst2) > 1:
            return lst2[0:1] + merge_list(lst1, lst2[1:])
        else:
            return lst2[0:1] + lst1

def func(lst : list[int], k : int) -> list[int]:
    if len(lst) == 2:
        if lst[0] <= lst[1]:
            return [lst[1], lst[0]]
        else:
            return lst
    if len(lst) == 1:
        return lst
    
    n = len(lst)
    
    if len(lst) > 2:
        a = func(lst[0:int(n/2)], 2)
        b = func(lst[int(n/2) : n], 2)
        res_fin = merge_list(a,b)
        return res_fin
    
    
    
# Optimized Merge logic - O(n log n)
def merge_sorted_lists(l1: list[int], l2: list[int]) -> list[int]:
    res = []
    i = j = 0
    # Use pointers (i, j) instead of slicing to save memory
    while i < len(l1) and j < len(l2):
        if l1[i] >= l2[j]:
            res.append(l1[i])
            i += 1
        else:
            res.append(l2[j])
            j += 1
    # Append remaining elements
    res.extend(l1[i:])
    res.extend(l2[j:])
    return res

def merge_sort_descending(lst: list[int]) -> list[int]:
    if len(lst) <= 1:
        return lst
    
    mid = len(lst) // 2
    left = merge_sort_descending(lst[:mid])
    right = merge_sort_descending(lst[mid:])
    
    return merge_sorted_lists(left, right)
```

```python
# product of all elements in the list except the current one
# O(n) in time

def productExceptSelf(nums: list[int]) -> list[int]:

    n = len(nums)
    prefix = []
    suffix = []
    results = []
    a = 1
    b = 1
    
    for i in range(n):
        a = a * nums[i]
        b = b * nums[n - 1 - i]
        prefix.append(a)
        suffix.append(b)
        results.append(0)

    for i in range(n):
        if i != 0 and i != n - 1:
            results[i] = prefix[i-1] * suffix[n-i-2]
        if i == 0:
            results[i] = suffix[n - 2]
        if i == n-1:
            results[i] = prefix[n - 2]

    return results


# Better solution in O(1) space and O(n) time

def productExceptSelf(nums: list[int]) -> list[int]:

    n = len(nums)
    res = [1 for _ in range(n)]

    for i in range(1,n):
        res[i] = res[i - 1] * nums[i - 1]

    a = 1
    for i in range(n - 1, -1, -1):
        res[i] = res[i] * a
        a = a*nums[i]

    return res

# example
nums = [1,2,3,4]
print(f"result{productExceptSelf(nums)}")
```

```python
def longestConsecutive(nums: list[int]) -> int:
    
    if len(nums) == 0:
        ml = 0
    else:
        res = {nums[i] : nums[i] for i in range(len(nums))}
        start = []

        for elem in nums:
            if elem + 1 in res:
                res[elem + 1] = elem
            elif elem - 1 in res:
                res[elem] = elem - 1
            else:
                try:
                    del res[elem]
                except:
                    continue
                
        for elem in res.keys():
            if res[elem] == elem:
                start.append(elem)        

        ml = 1
        for elem in start:
            cnt = 1
            while elem + 1 in res:
                cnt+=1
                elem += 1
            if cnt > ml:
                ml = cnt

    return ml
                
```

```python

def twoSum(nums: list[int], target: int) -> list[int]:
    idx1 = 0
    idx2 = 0

    for i in range(len(nums)):
        if target - nums[i] in nums[i+1:]:
            idx1 = i
            break
    for i in range(idx1 + 1, len(nums)):
        if target - nums[idx1] == nums[i]:
            idx2 = i
            break
            
    return [idx1, idx2]
```
