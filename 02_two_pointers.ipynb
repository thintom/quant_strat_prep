```python
def threeSum(self, nums: list[int]) -> list[list[int]]:
        nums = sorted(nums)
        res_fin = {}
        res = {nums[i]:i for i in range(len(nums))}

        for i in range(len(nums)):
            if (i > 1 and nums[i] == nums[i - 1]):
                continue
            else:
                target = nums[i]
                tmp_dict = {}
                tmp_dict[i] = target
                
                for j in range(len(nums)):
                    if i!= j:
                        a = - target - nums[j]
                        tmp_dict[j] = nums[j]
                        
                        if a in res and res[a] not in tmp_dict:
                            res_fin[ '*'.join(str(elem) for elem in sorted([target,nums[j],a]))] = 0
                            # res_fin.append([target, nums[j], a])
        res_fin = [[int(x) for x in elem.split('*')] for elem in res_fin]
        return res_fin

def threeSum(nums: list[int]) -> list[list[int]]: # faster algorithm
        nums = sorted(nums)
        res = []

        for i in range(len(nums)):
            if (i > 0 and nums[i] == nums[i - 1]):
                continue
            else:
                left = i + 1
                right = len(nums) - 1
                
                while left < right:
                    total = nums[i] + nums[left] + nums[right]
                    #print(left, right, total)
                    
                    if total == 0: #  and i != left and i!=right
                        res.append([nums[i], nums[left], nums[right]])

                        while left < right and nums[left] == nums[left + 1]:
                            left +=1
                        
                        while left < right and nums[right] == nums[right - 1]:
                            right -=1

                        left +=1
                        right-=1
                        
                    elif total < 0:
                        left += 1
                    else:
                        right -= 1
        return res
```

```python
def fourSum(nums: list[int], target: int) -> list[list[int]]:
        nums.sort()
        res = []

        for i in range(len(nums)):
            if (i > 0 and nums[i] == nums[i - 1]):
                continue
            else:
                j = i + 1
                while j < len(nums):
                    if (j > i + 1 and nums[j] == nums[j - 1]):
                        j+=1
                        continue
                    else:
                        cible = target - nums[i] - nums[j]
                        
                        left = j + 1
                        right = len(nums) - 1
                        
                        while left < right:
                            total = cible - nums[left] - nums[right]
                            
                            if total == 0:
                                res.append([nums[i] , nums[j], nums[left] , nums[right]])
                                
                                while left < right and nums[left] == nums[left + 1]:
                                    left +=1
                                        
                                while left < right and nums[right] == nums[right - 1]:
                                    right -=1
                                    
                                left +=1
                                right -=1
                                
                            if total < 0:
                                right -=1
                            if total>0:
                                left += 1
                        j += 1
        return res
```

```python
def maxArea(height: list[int]) -> int:
        res = {i:height[i] for i in range(len(height))}

        res_asc = {k:v for k,v in sorted(res.items(), key = lambda item: item[1])}
        i_list = list(res_asc.keys())
        h_list = list(res_asc.values())

        res_bis = {}
        cur_max = len(i_list) - 1
        cur_min = 0
        area_max = 0

        for i in range(len(i_list)):

            while cur_max in res_bis and cur_max > cur_min:
                cur_max -= 1
            
            while cur_min in res_bis and cur_min < cur_max:
                cur_min += 1
            
            if cur_max not in res_bis and cur_min not in res_bis:
                area1 = h_list[i] * (cur_max - i_list[i])
                area2 = h_list[i] * (i_list[i] - cur_min)
                area = max(area1, area2)
                

            if area > area_max:
                area_max = area
            
            res_bis[i_list[i]] = 0

        return area_max
```
